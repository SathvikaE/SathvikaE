```python
import cv2
import mediapipe as mp
import numpy as np

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

hands = mp_hands.Hands(static_image_mode=False,max_num_hands=1,min_detection_confidence=0.7,min_tracking_confidence=0.7)
from google.colab import files
uploaded = files.upload()
video_path= list(uploaded.keys())[0]
cap=cv2.VideoCapture(video_path)

output_file = 'tremor_detection_video.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out=  (cv2.VideoWriter(output_file, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4)))))

prev_wrist_position = None
tremor_threshold = 0.2
tremor_intensity=0

while cap.isOpened():
    success, image = cap.read()
    if not success:
        break

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    results = hands.process(image_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(image, hand_landmarks,mp_hands.HAND_CONNECTIONS)

            current_wrist_position = (hand_landmarks.landmark[mp_hands.HandLandmark.WRIST].x,
                                  hand_landmarks.landmark[mp_hands.HandLandmark.WRIST].y)
            if prev_wrist_position is not None:
               movement = np.sqrt((current_wrist_position[0] - prev_wrist_position[0]) ** 2 +
                               (current_wrist_position[1] - prev_wrist_position[1]) ** 2)

               if movement > tremor_threshold:
                    tremor_intensity+= movement
                    cv2.putText(image, "Tremor Detected", (50,50),cv2.FONT_ITALIC, 1, (0,0,255),2,cv2.LINE_AA)

    prev_wrist_position= current_wrist_position
    out.write(image)

cap.release()
out.release()
cv2.destroyAllWindows()
from google.colab import files
files.download(output_file)
```
